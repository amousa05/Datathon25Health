```python
import pandas as pd

# 1. Load the raw file
geo_path = "Ministry_of_Health_Service_Provider_Locations.csv"
moh = pd.read_csv(geo_path)

# 2. Show columns + first few rows
print(moh.head())         

# 3. Rename X, Y â†’ LONGITUDE, LATITUDE
moh = moh.rename(columns={
    "X": "LONGITUDE",
    "Y": "LATITUDE",
    "ADDRESS_LINE_1" : "ADDRESS"
})

# 4. Keep important columns only
cols_to_keep = ["LONGITUDE", "LATITUDE", "ENGLISH_NAME", "SERVICE_TYPE", "ADDRESS"]
moh = moh[cols_to_keep]

# 5. Filter to hospitals only
moh_hospitals = moh[moh["SERVICE_TYPE"] == "Hospital"].copy()



#save cleaned file for merging later
moh_hospitals.to_csv("Ministry_of_Health_Hospitals_Geo.csv", index=False)
```
